```python
sample1 = newdata[400000:]
sample1
```

```python
sample1.isna().sum()
```

```python
tweets = pyfunctional.seq(sample1['Cleaned_tweets']).map(str.strip)
tweets[0]
```

```python
def predictor(sentence):
    try:
        sentiment_model.predict(sentence)
    except Exception:
        print("This sentence is bad: ", sentence)
        return sentence
    return sentence
```

```python
def splitor(predicted_sentence):
    try: 
        return predicted_sentence.labels[0].score, predicted_sentence.labels[0].value
    except Exception:
        print("This is the empty sentence: ", predicted_sentence)
        return "BAD SENTENCE", "BAD SENTENCE"
```

```python
flaired_predictions = tweets.map(flair.data.Sentence).map(predictor).map(splitor)
flaired_predictions
```

```python
probabilities = flaired_predictions.map(
    lambda flaired_prediction: flaired_prediction[0]).to_list()
sentiment = flaired_predictions.map(
    lambda flaired_prediction: flaired_prediction[1]).to_list()
```

```python
sample1['probability'] = probabilities
sample1['sentiment'] = sentiment
```
